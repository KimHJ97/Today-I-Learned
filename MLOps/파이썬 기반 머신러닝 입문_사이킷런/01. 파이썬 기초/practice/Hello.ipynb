```python
# print
message = 'Python'
print("Hello {message}") # 기본 출력
print(f"Hello {message}") # String 포맷 출력: f 키워드 사용
print("Hello {}".format(message))
print("Hello %s" % (message))

# List
# append(), remove(), insert(), index()
print("======================= List ======================= ")
aList = []
bList = list()

aList.append(1)
aList.append(2)
print(aList)
aList.remove(1)
print(aList)
aList.insert(0, 1)
print(aList)
print(aList.index(1))

# Dict
# keys(), values(), items()
print("======================= Dict ======================= ")
d = {}
d['국어'] = 90
d['영어']= 80
d['수학'] = 100
print(d)
del d['수학']
print(d)
print(d.keys())
print(d.values())
print(d.items())

# Tuple
print("======================= Tuple ======================= ")
t = (1,)
print(t)
```

```python
# Iterator
print("======================= Iterator ======================= ")
a = [4, 2]
a.append('a')
a.append('python')
print(a[2:4]) # 범위 검색
print(a[:2]) # 첫번쨰 ~ 1번
print(a[2:]) # 2번 ~ 마지막
print(a[:]) # 전체

# Copy
print("======================= Copy ======================= ")
print("None Copy")
a = [1,2,3,4]
b = a
b.append(5)
b.append(6)
print(a)

print("Copy")
a = [1,2,3,4]
b = a[:]
b.append(5)
b.append(6)
print(a)

# string <-> list
# index(), list(), splite(), join(), replace()
print("======================= string <-> list ======================= ")
str = 'I like python'
print(str.split()) # 기본적으로 공백 문자를 구분자로 사용한다.
print(str.split(' '))
print(" ".join(str.split()))

```

```python
# 순환문
print("======================= 순환문 ======================= ")
print("===> range(1,5)")
for i in range(1, 5):
    print(i)

print("===> range(1,5, 2)")
for i in range(1, 5, 2):
    print(i)

print("===> char in 'hello'")
for char in "hello":
    print(char)

print("===> i in range(len(['국어', '영어', '수학']))")
subjects = ['국어', '영어', '수학']
for i in range(len(subjects)):
    print(subjects[i])

print("===> key, values in scores.items()")
scores = {"국어": 100, "영어": 90, "수학": 80}
for key, value in scores.items():
    print(f"key:{key}, value:{value}")
```

```python
def extract_number(a):
    temp = []
    while(1):
        q = a // 10
        r = a % 10
        a = q
        temp.append(r)
        if a < 10:
            temp.append(a)
            break

    return temp

numbers = extract_number(12345)
print(numbers)

numbers.reverse()
print(numbers)

reverseNumbers2 = list(reversed(numbers))
print(reverseNumbers2)
```
